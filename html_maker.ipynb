```python
import texttable as tt

# Define the hyperlinks
links = {
    "choose_ide": '<a href="IDE_choice.html">Choosing an IDE</a>',
    "vscode_link": 'VS Code, etc.',
    "programming_languages": 'Python, MATLAB, Julia, etc.',
    "scientific_computing": '<a href="OOP.html">Scientific Computing</a>',
    "scientific_tools": 'matplotlib, sklearn, pandas, numpy, etc.',
    "ml_packages": '<a href="OOP.html">Machine Learning Packages</a>',
    "ml_tools": 'Python - TensorFlow, Keras, sklearn\nMATLAB - Deep Learning Toolbox',
    "ml": '<a href="OOP.html">Machine Learning/AI</a>',
    "ml_general": 'General, types of ML, math Fundamentals, etc.',
    "other_stuff": 'Other Stuff',
    "other_details": 'Productivity tool, HPC, CLI, deployment, etc.'
}

# Create tables for each section using texttable
def create_table(rows, width):
    tab = tt.Texttable(max_width=width)
    tab.set_cols_align(["c"] * len(rows[0]))
    tab.set_cols_valign(["m"] * len(rows[0]))
    tab.add_rows(rows)
    return tab.draw()

# Define the tables
table1 = create_table([
    [links["choose_ide"]],
    [links["vscode_link"]]
], width=40)

table2 = create_table([
    ["Step 2: Using a programming language"],
    [links["programming_languages"]]
], width=40)

table3 = create_table([
    [links["scientific_computing"]],
    [links["scientific_tools"]]
], width=40)

table4 = create_table([
    [links["other_stuff"]],
    [links["other_details"]]
], width=40)

table5 = create_table([
    [links["ml_packages"]],
    [links["ml_tools"]]
], width=40)

table6 = create_table([
    [links["ml"]],
    [links["ml_general"]]
], width=40)

# Combine tables into the ASCII art
ascii_art = f"""
+----------------------------------+         +---------------------------------+         +---------------------------------+
|             Choosing             |         |            Basics of            |         |           Scientific            |
|              an IDE              |         |        programming language     |         |           computing             |
+==================================+ ---- >  +=================================+ ---- >  +=================================+
|           VS Code, etc.          |         |  Python, MATLAB, Julia, etc.    |         |                                 |
+----------------------------------+         +---------------------------------+         +---------------------------------+ 
												         |
													 |
													 |
													 V
+----------------------------------+         +---------------------------------+         +-------------------------------------+
|              Other               |         |           Popular ML            |         |         Machine learning &          |
|              stuff               |         |            packages             |         |       artificial intelligence       |
+==================================+ < ----  +=================================+ < ----  +=====================================+
|Productivity, HPC, CLI, deployment|         |      Keras, sklearn, etc.       |         | General, types, math fundamentals   |
+----------------------------------+         +---------------------------------+         +-------------------------------------+ 
"""

# Define the HTML structure
html_content = f"""
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Education Resources</title>
    <style>
        body {{
            font-family: monospace;
        }}
        .container {{
            display: flex;
            flex-direction: column;
            align-items: center;
        }}
        .ascii-art {{
            white-space: pre;
            font-size: 12px;
        }}
    </style>
</head>
<body>
    <div class="container">
        <h1>Welcome to my website</h1>
        <p>I have a list of education resources I like. I also have a list of my publications.</p>
        
        <h1 id="education-resources">Education Resources</h1>
        <pre class="ascii-art">
{ascii_art}
        </pre>
        
        <h1 id="publications">Publications</h1>
        <p>If you'd like a copy of any of my publications, send me an email.</p>
        
        <h2>Published</h2>
        <p><a href="https://etd.ohiolink.edu/acprod/odb_etd/etd/r/1501/10?clear=10&p10_accession_num=osu1679258588528361">2023 MS Thesis - Assessment of Multi-Fidelity Surrogate Methods for Expedient Loads Prediction in High-Speed Flows</a></p>
        <p><a href="https://doi.org/10.2514/6.2023-3844">2023 AIAA Aviation Conference Paper - Assessment of Multi-Fidelity Surrogate Approaches for Expedient Loads Prediction in High-Speed Flows</a></p>
        
        <h2>In Peer Review</h2>
        <p>Assessment of Multi-Fidelity Surrogate Approaches for Expedient Loads Prediction in High-Speed Flows, Journal of Spacecraft and Rockets</p>
        <p>The Impact of Upstream Static Deformation on Flow Past a Cylinder/Flare, MDPI-Aerospace</p>
    </div>
</body>
</html>
"""

# Save the HTML content to a file named "index.html"
with open("index.html", "w") as file:
    file.write(html_content)

print(html_content)
print("HTML content has been written to index.html")
```
